## Reading the Dataset

```python
import pandas as pd
df = pd.read_csv('BX-Books.csv',error_bad_lines=False,encoding='latin-1',sep=';')
df.head()
```

```python
#list variables
df.info()
```

Pre-processing Data to Build the Recommendation System

```python
#Removing Duplicates
df.duplicated(subset='Book-Title').sum() 

# 29,225 it means there are this much dublicated book titles in the dataframe
```

```python
#TO eleminate them we will use 
df = df.drop_duplicates(subset='Book-Title')
```

```python
#To make sure that we removed dublicated book titles
df.duplicated(subset='Book-Title').sum() # 0
```

## Random Sampling

```python
# we use sample of 15000 rows to avoid running into memory errors
sample_size = 15000
df = df.sample(n=sample_size, replace=False, random_state=490)

df = df.reset_index()
df = df.drop('index',axis=1)
```

## Processing Text Data

```python
#let see the head of the data frame
df.head()
```

## Data Cleaning

```python
#here we are removing the whitespaces from the Book-Author that the CountVectorizer should not count the first and last name as separate
def clean_text(author):
    result = str(author).lower()
    return(result.replace(' ',''))

df['Book-Author'] = df['Book-Author'].apply(clean_text)
```

```python
df.head()
```

## Now, let’s convert the book title and publisher to lowercase:

```python
df['Book-Title'] = df['Book-Title'].str.lower()
df['Publisher'] = df['Publisher'].str.lower()
```

```python
#All in all, we will combine these variables to create a single variables
df2 = df.drop(['ISBN','Image-URL-S','Image-URL-M','Image-URL-L','Year-Of-Publication'],axis=1)

df2['data'] = df2[df2.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)

print(df2['data'].head())
```

```python
#Now we can use Scikit-Learn's CountVectorizer() on the combined text data
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

## Building the Recommendation System

```python
#
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df = pd.DataFrame(similarities, columns=df['Book-Title'], index=df['Book-Title']).reset_index()
df.head()
```

## Displaying User Recommendations

```python
#here we can print the list 10 similar books as a recomendation
input_book = 'enemies (carrier action adventures, 15)'
recommendations = pd.DataFrame(df.nlargest(11,input_book)['Book-Title'])
recommendations = recommendations[recommendations['Book-Title']!=input_book]
print(recommendations)
```
